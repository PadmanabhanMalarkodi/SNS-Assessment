```python
# If it is a range we can take the average
# If the unit is in meter or perch or any other we are ignoring it.

def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df3 = df2.copy()
df3['total_sqft'] = df3['total_sqft'].apply(convert_sqft_to_num)
df3
```

### i) Non-numerical column
#### There so many null values in society column, Inorder to prevent the loss of data I'm removing that column for my analysis.

```python
df4 = df3.drop(['society'], axis = 1)
df4.head()
```

```python
df4.isnull().sum()
```

```python
df4 = df4.dropna(subset=['location'], axis=0)  # Drops rows where 'location' is null
df4.isnull().sum()
```

### ii) Numerical columns

```python
# Dropping rows where 'size' is null
df4 = df4.dropna(subset=['size'], axis=0)

# Replacing missing values in 'bath' column with the mode
df4['bath'] = df4['bath'].fillna(df4['bath'].mode()[0])

# Replacing missing values in 'total_sqft' column with the mean
df4['total_sqft'] = df4['total_sqft'].fillna(df4['total_sqft'].mean())

# Replacing missing values in 'balcony' column with the mean
df4['balcony'] = df4['balcony'].fillna(df4['balcony'].mean())

# Checking for remaining missing values
print(df4.isnull().sum())
```

```python
df5 = df4.copy()

df5['bhk'] = df5['size'].apply(lambda x: int(x.split(' ')[0]))
df5
```
